```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
# 공부시간 X와 성적 Y의 리스트를 만들기
data = [[2, 81], [4, 93], [6, 91], [8, 97]]
x = [i[0] for i in data]
y = [i[1] for i in data]
```

```python
# 그래프로 나타내기
plt.figure(figsize = (8, 5))
plt.scatter(x, y)
plt.show()
```

```python
# 리스트로 되어 있는 x와 y값을 넘파이 배열로 바꾸기(인덱스를 주어 하나씩 불러와 계산이 가능하게 하기 위함)
x_data = np.array(x)
y_data = np.array(y)
```

```python
# 리스트 a와 절편 b의 값 초기화
a = 0
b = 0
```

```python
# 학습률 정하기
lr = 0.05
```

```python
# 몇 번 반복될지 설정(0부터 세므로 원하는 반복횟수에 +1)
epochs = 2001
```

```python
# 경사하강법 시작
for i in range(epochs): # 에포크 수만큼 반복
    y_pred = a * x_data + b  
    error = y_data - y_pred 
    # 오차함수를 a로 미분한 값
    a_diff = -(1/len(x_data)) * sum(x_data * (error))
    # 오차함수를 b를 미분한 값
    b_diff = -(1/len(x_data)) * sum(y_data - y_pred)
    
    a = a - lr * a_diff 
    b = b - lr * b_diff
    
    if i % 100 == 0: 
        print('epoch=%.f, 기울기=%.04f, 절편=%.04f' % (i, a, b))
    
# 앞서 구한 기울기와 절편을 이용해 그래프를 다시 그리기
y_pred = a * x_data + b
plt.scatter(x, y)
plt.plot([min(x_data), max(x_data)], [min(y_pred), max(y_pred)])
plt.show()
```

```python
# 다중 선형 회귀
data = [[2, 0, 81], [4, 4, 93], [6, 2, 91], [8, 3, 97]]
x1 = [i[0] for i in data]
x2 = [i[1] for i in data]
y = [i[2] for i in data]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
```

```python
ax = plt.axes(projection = '3d') # 그래프 유형 정하기
ax.set_xlabel('study_hours')
ax.set_ylabel('private_class')
ax.set_zlabel('Score')
ax.dist = 11
ax.scatter(x1, x2, y)
plt.show()
```

```python
# 리스트로 되어 있는 x와 y값을 넘파이 배열로 바꾸기(인덱스로 하나씩 불러와 계산할 수 있도록 하기 위함)
x1_data = np.array(x1)
x2_data = np.array(x2)
y_data = np.array(y)
```

```python
# 기울기 a와 절편 b의 값 초기화
a1 = 0
a2 = 0
b = 0
```

```python
# 학습률
lr = 0.05
```

```python
# 몇 번 반복할지 설정(0부터 세므로 원하는 반복횟수에 +1)
epochs = 2001
```

```python
# 경사하강법 시작
for i in range(epochs): # epoch 수 만큼 반복
    y_pred = a1 * x1_data + a2 * x2_data + b # y를 구하는 식 구하기
    error = y_data - y_pred 
    # 오차 함수를 a1으로 미분한 값
    a1_diff = -(1/len(x1_data)) * sum(x1_data * (error))
    # 오차 함수를 a2으로 미분한 값
    a2_diff = -(1/len(x2_data)) * sum(x2_data * (error))
    # 오차함수를 b로 미분한 값
    b_diff = -(1/len(x1_data)) * sum(y_data - y_pred)
    a1 = a1 - lr * a1_diff # 학습률을 곱해 기존의 a1 값 업데이트
    a2 = a2 - lr * a2_diff # 학습률을 곱해 기존의 a2 값 업데이트
    b = b - lr * b_diff # 학습률을 곱해 기존의 b값 업데이트
    
    if i % 100 == 0: # 100번 반복될 때마다 현재의 a1, a2, b값 출력
        print('epoch=%.f, 기울기1=%.04f, 기울기2=%.04f, 절편=%.04f' %(i, a1, a2, b))
```
